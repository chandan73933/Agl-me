```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from apyori import apriori
```

```python
df =pd.read_csv("Desktop/customerdata.csv")
```

```python
final=df.groupby(['canonical_id','call_type','industry_std_desc','model',]).size().reset_index().head(1000)
```

```python
final.head()
```

```python
final.shape
```

```python
num_records= len(final)
print(num_records)
```

```python
records = []
for i in range(0,1000):
    records.append([str(final.values[i,j]) for j in range(0,4)])
```

```python
#Building model
association_rules = apriori(records, min_support=0.0056,min_confidence=0.2,min_lift=3,min_length=2)
association_results=list(association_rules)
```

```python
print(len(association_results))
```

```python
print(association_results[0])
```

```python
results=[]
for item in association_results:
    pair = item[0]
    items = [x for x in pair]
    
    value0 = str(items[0])
    value1 = str(items[1])
    value2 = str(item[1])[:5]
    value3 = str(item[2][0][2])[:5]
    value4 = str(item[2][0][3])[:5]
    
    rows = (value0,value1,value2,value3,value4)
    
    results.append(rows)
    
    Label = ['Title1','Title2','Support','Confidence','Lift']
    
    store_suggestion = pd.DataFrame.from_records(results,columns=Label)
    #final.sort_values(by ='Lift', ascending = False, inplace = True)
    print(store_suggestion)
   
```

```python
for item in association_results:

    # first index of the inner list
    # Contains base item and add item
    pair = item[0] 
    items = [x for x in pair]
    print("Association_Rule: " + items[0] + " -> " + items[1])

    #second index of the inner list
    print("Support: " + str(item[1]))

    #third index of the list located at 0th
    #of the third index of the inner list

    print("Confidence: " + str(item[2][0][2]))
    print("Lift: " + str(item[2][0][3]))
    print("=====================================")
```
